### 7.c - Compute statistics by group

Compute the proportion of female vs. male survivors of the RMS Titanic.

```python
titanic[['Sex', 'Survived']].groupby('Sex').mean()
```

Now compute the proportion of female vs. male survivors of the RMS Titanic, *along with the **s**tandard **e**rror of the **m**ean*. The **bold** type should give you a hint about the name of the method to compute the standard error. To do this, you'll need to combine the `groupby()` and `agg()` methods!

```python
stats = {'Survived' : ['mean', 'sem']}
titanic[['Sex', 'Survived']].groupby('Sex').agg(stats)
```

What does this tell you about gender roles when the RMS Titanic was sunk?

*The proportion of women who survived the sinking of the titanic was much larger than the proportion of men who survived. This tells us that gender roles at the time valued women's survival from sinking ships over men's.*

Compute the proportion of survivors by cabin class and their standard error.

```python
stats = {'Survived' : ['mean', 'sem']}
titanic[['Pclass', 'Survived']].groupby('Pclass').agg(stats)
```

What does this tell you about socio-economic status when the RMS Titanic was sunk?

*The proportion of first class passengers that survived the sinking of the Titanic was much higher than the proportion of survivors from any other class. This tells that people of higher socio-economic status were valued more/given a better opportunity to escape than those of lwoer socio-economic status.*
